```python
import cv2
import os

def extract_frames_with_priority_deletion(video_path, output_dir, target_frames=60):
    """
    Extract frames focusing on the middle section of the video.
    If the video has fewer than 60 frames, copy the last frame to fill the deficit.
    If the video has more than 60 frames, delete all extra frames from the end.

    :param video_path: Path to the video file.
    :param output_dir: Directory where the frames will be saved.
    :param target_frames: The number of frames to extract (default 60).
    """
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the total number of frames in the video
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Read and store all frames
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # If the total number of frames is less than the target, copy the last frame to fill the gap
    if total_frames < target_frames:
        for i in range(total_frames):
            frame_filename = os.path.join(output_dir, f"{i}.png")  # Save as .png starting from 0.png
            cv2.imwrite(frame_filename, frames[i])

        # Copy the last frame to fill the deficit
        last_frame = frames[-1]
        for i in range(total_frames, target_frames):
            frame_filename = os.path.join(output_dir, f"{i}.png")
            cv2.imwrite(frame_filename, last_frame)

        print(f"Copied last frame to fill the deficit for {video_path}.")
        cap.release()
        return

    # If the video has more than the target number of frames, delete all extra frames from the end
    if total_frames > target_frames:
        middle_start = 0
        middle_end = target_frames
    else:
        middle_start = max((total_frames // 2) - int(target_frames // 1.5), 0)
        middle_end = min(middle_start + target_frames, total_frames)

    extracted_frame_count = 0

    # Iterate through the middle section and save frames
    for i in range(middle_start, middle_end):
        frame_filename = os.path.join(output_dir, f"{extracted_frame_count}.png")  # Save as .png starting from 0.png
        cv2.imwrite(frame_filename, frames[i])
        extracted_frame_count += 1

    cap.release()
    print(f"Extracted {extracted_frame_count} frames from {video_path}")

# Main function to handle the directory structure for multiple folders
def process_multiple_video_folders(video_folders, output_base_folder):
    for video_folder in video_folders:
        # Extract the set name (e.g., "set_1" or "set_11") from the folder name
        set_name = os.path.basename(video_folder)
        output_set_folder = os.path.join(output_base_folder, set_name)

        # Go through all ".mp4" video files in the current set folder
        for video_file in os.listdir(video_folder):
            if video_file.endswith('.mp4'):  # Only process .mp4 files
                video_path = os.path.join(video_folder, video_file)

                # Generate the output directory path without additional suffixes
                word_name = os.path.splitext(video_file)[0]  # Get the word name without extension
                word_output_dir = os.path.join(output_set_folder, word_name)

                # Process the video and save frames
                print(f"Processing {video_file} in {set_name}...")
                extract_frames_with_priority_deletion(video_path, word_output_dir, target_frames=60)

# Example usage:
video_folders = [
    r'D:\PycharmProjects\pro_dis_2\videos\set_1',  # Path to "set_1"
    r'D:\PycharmProjects\pro_dis_2\videos\set_11'  # Path to "set_11"
]
output_base_folder = r'D:\PycharmProjects\pro_dis_2\collected_data\actual_datasets'  # Updated base folder path

# Call the processing function for all video folders
process_multiple_video_folders(video_folders, output_base_folder)
```

```python
import os
import cv2
import numpy as np
import random

from concurrent.futures import ThreadPoolExecutor, as_completed

# Define source and output directories
source_dirs = [
    r'D:\PycharmProjects\pro_dis_2\collected_data\actual_datasets\set_1',
    r'D:\PycharmProjects\pro_dis_2\collected_data\actual_datasets\set_11'
]
output_base_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\augmented_datasets'

# Set the naming for the output sets
output_sets = {
    source_dirs[0]: [f"set_{i}" for i in range(2, 11)],  # Sets 2 to 10
    source_dirs[1]: [f"set_{i}" for i in range(12, 21)]  # Sets 12 to 20
}

# ImageDataGenerator for basic augmentations
datagen = ImageDataGenerator(
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    # rotation_range=5,
    # zoom_range=[0.8, 1.2],
    # width_shift_range=0.1,
    # height_shift_range=0.1,
    # shear_range=0.2
)

# Function to apply noise to an image
def add_noise(image):
    row, col, ch = image.shape
    mean = 0
    sigma = 0.1
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    noisy_image = np.clip(image + gauss * 255, 0, 255).astype(np.uint8)
    return noisy_image

# Function to downsample (simulate lower resolution)
def downsample_image(image, scale=0.5):
    h, w = image.shape[:2]
    image = cv2.resize(image, (int(w * scale), int(h * scale)))
    return cv2.resize(image, (w, h))

# Function to apply blur
def apply_blur(image, kernel_size=3):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

# Function to augment a single frame
def augment_frame(image_path, output_dir, frame_idx):
    img = cv2.imread(image_path)
    img = img_to_array(img)

    # Apply basic augmentations
    img = datagen.random_transform(img)

    # Additional augmentations
    if random.random() < 0.3:
        img = add_noise(img)
    if random.random() < 0.2:
        img = downsample_image(img)
    if random.random() < 0.2:
        img = apply_blur(img)

    # Convert to image format and save
    aug_img = array_to_img(img)
    aug_img.save(f"{output_dir}/{frame_idx}.png")

# Function to process each frame in a folder with parallel execution
def process_frame_folder(source_word_dir, output_word_dir):
    frames = sorted(os.listdir(source_word_dir))
    num_frames = len(frames)

    # Process frames in parallel using ThreadPoolExecutor
    with ThreadPoolExecutor() as executor:
        futures = [
            executor.submit(augment_frame, os.path.join(source_word_dir, frame), output_word_dir, idx)
            for idx, frame in enumerate(frames)
        ]

        # Ensuring all frames complete without any missing saves or skipped frames
        for future in as_completed(futures):
            future.result()

# Main loop to generate augmented datasets with folder-wise parallel processing
for source_dir in source_dirs:
    # Print statement to indicate which set is currently being processed
    if source_dir.endswith('set_1'):
        print("Starting processing for set_1...")
    elif source_dir.endswith('set_11'):
        print("Starting processing for set_11...")

    for output_set_name in output_sets[source_dir]:
        output_set_dir = os.path.join(output_base_dir, output_set_name)
        os.makedirs(output_set_dir, exist_ok=True)

        # Process each word directory within the source set
        for word_dir in os.listdir(source_dir):
            source_word_dir = os.path.join(source_dir, word_dir)
            output_word_dir = os.path.join(output_set_dir, word_dir)
            os.makedirs(output_word_dir, exist_ok=True)

            if os.path.isdir(source_word_dir):
                print(f"Augmenting frames for {output_set_name}/{word_dir}...")
                process_frame_folder(source_word_dir, output_word_dir)

print("Augmented datasets created successfully in 'augmented_datasets_uneven_frames'.")

# Validation step to check for 60 frames in each word folder under each set
all_ok = True  # Flag to track if any folder is missing frames

for output_set_name in os.listdir(output_base_dir):
    set_dir = os.path.join(output_base_dir, output_set_name)

    # Check each word folder within the set
    for word_folder in os.listdir(set_dir):
        word_folder_path = os.path.join(set_dir, word_folder)

        if os.path.isdir(word_folder_path):
            frame_count = len(os.listdir(word_folder_path))

            # Check if the word folder has exactly 60 frames
            if frame_count != 60:
                print(f"Warning: {output_set_name}/{word_folder} has {frame_count} frames instead of 60.")
                all_ok = False

# Final positive message if all folders have exactly 60 frames
if all_ok:
    print("All sets and word folders are correctly augmented with exactly 60 frames each. Validation successful!")
```

```python
import os
import shutil

# Define source and destination directories
actual_datasets_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\actual_datasets'
augmented_datasets_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\augmented_datasets'
complete_datasets_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\complete_datasets'

# List of all source directories to copy from
source_dirs = [
    actual_datasets_dir,  # Contains set_1 and set_11
    augmented_datasets_dir  # Contains set_2 to set_10 and set_12 to set_20
]

# Create the complete_datasets directory if it doesn't exist
os.makedirs(complete_datasets_dir, exist_ok=True)

# Function to copy all set folders and their contents
def copy_sets(source_base_dir, destination_base_dir):
    for set_folder in os.listdir(source_base_dir):
        set_folder_path = os.path.join(source_base_dir, set_folder)
        dest_set_folder_path = os.path.join(destination_base_dir, set_folder)

        # Copy the set folder if it exists and is a directory
        if os.path.isdir(set_folder_path):
            print(f"Copying {set_folder} from {source_base_dir} to complete_datasets...")
            shutil.copytree(set_folder_path, dest_set_folder_path, dirs_exist_ok=True)

# Copy sets from both actual_datasets and augmented_datasets to complete_datasets
for source_dir in source_dirs:
    copy_sets(source_dir, complete_datasets_dir)

print("All datasets have been successfully copied to 'complete_datasets'.")
```

```python
import os
import shutil

# Function to copy everything from the source directory to the destination directory
def copy_entire_directory(source_dir, destination_dir):
    # Copy the entire directory tree to the destination
    if os.path.exists(destination_dir):
        shutil.rmtree(destination_dir)  # Remove existing directory to ensure a fresh copy
    shutil.copytree(source_dir, destination_dir)
    print(f"Copied entire directory from {source_dir} to {destination_dir}")

# Function to rename word folders by appending the set number
# Function to rename word folders by appending the set number
def rename_word_folders(destination_dir):
    # Iterate through each set directory in the destination directory
    for set_folder in os.listdir(destination_dir):
        set_folder_path = os.path.join(destination_dir, set_folder)

        # Check if the item is a directory and matches the "set_x" format
        if os.path.isdir(set_folder_path) and set_folder.startswith("set_"):
            # Extract the set number "x" from "set_x"
            set_number = set_folder.split("_")[1]

            # Iterate through each word folder in the current set folder
            for word_folder in os.listdir(set_folder_path):
                word_folder_path = os.path.join(set_folder_path, word_folder)

                # Check if the item is a directory (word folder)
                if os.path.isdir(word_folder_path):
                    # Create the new folder name with the set number as a suffix
                    new_word_folder_name = f"{word_folder}_{set_number}"
                    new_word_folder_path = os.path.join(set_folder_path, new_word_folder_name)

                    # Rename the word folder using shutil.move()
                    shutil.move(word_folder_path, new_word_folder_path)
                    print(f"Renamed {word_folder_path} to {new_word_folder_path}")

# Function to move all word folders to the root of the destination directory and delete the set_x folders
def move_word_folders(destination_dir):
    # Iterate through each set directory in the destination directory
    for set_folder in os.listdir(destination_dir):
        set_folder_path = os.path.join(destination_dir, set_folder)

        # Check if the item is a directory and matches the "set_x" format
        if os.path.isdir(set_folder_path) and set_folder.startswith("set_"):
            # Iterate through each word folder in the current set folder
            for word_folder in os.listdir(set_folder_path):
                word_folder_path = os.path.join(set_folder_path, word_folder)

                # Move the word folder to the root of the destination directory
                new_word_folder_path = os.path.join(destination_dir, word_folder)
                shutil.move(word_folder_path, new_word_folder_path)
                print(f"Moved {word_folder_path} to {new_word_folder_path}")

            # Remove the empty set_x folder after moving its contents
            os.rmdir(set_folder_path)
            print(f"Deleted empty folder: {set_folder_path}")

if __name__ == "__main__":
    # Define the source and destination directories
    source_dir = "D:/PycharmProjects/pro_dis_2/collected_data/processed_datasets_five"
    destination_dir = "D:/PycharmProjects/pro_dis_2/collected_data/complete_datasets_crop_six"

    # Step 1: Copy the entire source directory to the destination
    copy_entire_directory(source_dir, destination_dir)

    # Step 2: Rename the word folders by appending the set number
    rename_word_folders(destination_dir)

    # Step 3: Move all the word folders to the root of the destination directory and delete the set_x folders
    move_word_folders(destination_dir)

    print("All operations completed.")
```

```python
import os
import cv2
import json
import numpy as np
import imageio.v2 as imageio

# Path to the folder containing word folders with frames
word_folders_path = r'D:\PycharmProjects\pro_dis_2\collected_data\complete_datasets_crop_six'

# List to store all frames for each word sequence
all_words = []
labels = []

# Set FPS for video creation
fps = 30  # You can adjust this value as needed

# Iterate over each word folder
for word_folder in sorted(os.listdir(word_folders_path)):
    word_path = os.path.join(word_folders_path, word_folder)
    print(f'Processing: "{word_folder}"')

    # Check if the path is a directory (word folder)
    if os.path.isdir(word_path):
        word_frames = []

        # Iterate over each frame image in sequential order
        frame_files = sorted(
            [f for f in os.listdir(word_path) if f.endswith('.png')],
            key=lambda x: int(os.path.splitext(x)[0])
        )

        for frame_file in frame_files:
            frame_path = os.path.join(word_path, frame_file)

            # Read the frame image
            frame = cv2.imread(frame_path)
            if frame is not None:
                # Convert frame to a list of pixel values (height  width  channels)
                frame_data = frame.tolist()
                word_frames.append(frame_data)

        # Add the word frames to the all_words list and create data.txt and video.mp4
        if word_frames:
            all_words.append(word_frames)
            labels.append(word_folder)

            # Save word frames to a data.txt file in JSON format
            txt_path = os.path.join(word_path, "data.txt")
            with open(txt_path, "w") as f:
                f.write(json.dumps(word_frames))

            # Create video.mp4 using the frames in all_words without saving .png files
            video_path = os.path.join(word_path, "video.mp4")
            images = []

            for frame_data in word_frames:
                # Convert the frame data back to an array and add to the images list
                img_array = np.array(frame_data, dtype=np.uint8)
                images.append(img_array)

            # Save the video from the list of images
            imageio.mimsave(video_path, images, fps=fps)

            # Clear the all_words list for the next iteration
            all_words.clear()

print("\nData collection completed.")
print("The data.txt and video.mp4 files have been created in their corresponding word folders.")
```

```python
import os
import shutil
'_12', '_3', '_13'
def copy_selected_folders(source_dir, destination_dir, suffixes=('_3', '_12', '_13')):
    """
    Copy folders ending with specified suffixes from the source directory to the destination directory.

    :param source_dir: Path to the source directory containing word folders.
    :param destination_dir: Path to the destination directory where selected folders will be copied.
    :param suffixes: Tuple of suffixes to match folder names.
    """
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_dir, exist_ok=True)

    # Iterate over all items in the source directory
    for folder_name in os.listdir(source_dir):
        folder_path = os.path.join(source_dir, folder_name)

        # Check if the item is a folder and ends with any of the specified suffixes
        if os.path.isdir(folder_path) and folder_name.endswith(suffixes):
            # Destination path for the folder
            dest_path = os.path.join(destination_dir, folder_name)

            # Copy the entire folder to the destination directory
            shutil.copytree(folder_path, dest_path)
            print(f"Copied {folder_name} to {destination_dir}")

# Paths to the source and destination directories
source_directory = r'D:\PycharmProjects\pro_dis_2\collected_data\Rest_x_(complete_datasets_crop_six)'
destination_directory = r'D:\PycharmProjects\pro_dis_2\collected_data\complete_datasets_crop_six'

# Run the function
copy_selected_folders(source_directory, destination_directory)

print("\nAll selected folders have been copied.")
```

```python
import os

def rename_folders_suffix(source_dir, old_suffix='_11', new_suffix='_2'):
    """
    Rename all folders in the source directory that end with the specified old suffix to the new suffix.

    :param source_dir: Path to the source directory containing the folders to rename.
    :param old_suffix: The suffix to look for in folder names.
    :param new_suffix: The new suffix to replace the old suffix with.
    """
    for folder_name in os.listdir(source_dir):
        folder_path = os.path.join(source_dir, folder_name)

        # Check if the item is a folder and ends with the old suffix
        if os.path.isdir(folder_path) and folder_name.endswith(old_suffix):
            # New folder name with the updated suffix
            new_folder_name = folder_name[:-len(old_suffix)] + new_suffix
            new_folder_path = os.path.join(source_dir, new_folder_name)

            # Rename the folder
            os.rename(folder_path, new_folder_path)
            print(f"Renamed {folder_name} to {new_folder_name}")

# Path to the directory containing the folders to rename
source_directory = r'D:\PycharmProjects\pro_dis_2\collected_data\complete_datasets_crop_six'

# Run the function
rename_folders_suffix(source_directory)

print("\nAll folders ending with '_11' have been renamed to end with '_2'.")
```
