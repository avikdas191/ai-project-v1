**for copying videos**

```python
import os
import shutil

# Define source and destination directories
source_dir = r'D:\PycharmProjects\pro_dis_2\project_videos\old video'
destination_dir = r'D:\PycharmProjects\pro_dis_2\project_videos'

# List of specific word video filenames to copy (without the extension)
word_videos_to_copy = ['cup', 'eat', 'go', 'hot', 'sit']

# Loop through each set folder in the source directory
for set_folder in os.listdir(source_dir):
    set_folder_path = os.path.join(source_dir, set_folder)

    # Only proceed if it's a directory
    if os.path.isdir(set_folder_path):
        # Create the corresponding set folder in the destination
        destination_set_path = os.path.join(destination_dir, set_folder)
        os.makedirs(destination_set_path, exist_ok=True)
        print(f"\nCreated/checked set folder: {destination_set_path}")

        # Loop through each specified word video and copy it
        for word_video in word_videos_to_copy:
            video_filename = f"{word_video}.mp4"  # Add the extension to match the source file name
            video_path = os.path.join(set_folder_path, video_filename)
            if os.path.isfile(video_path):
                destination_video_path = os.path.join(destination_set_path, video_filename)
                shutil.copy(video_path, destination_video_path)
                print(f"Copied {video_filename} from {set_folder_path} to {destination_video_path}")
            else:
                print(f"{video_filename} not found in {set_folder_path}")
```

**for combined images**

```python
import os
import shutil

# Define source and destination directories
source_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\old data\3_combined_images'
destination_dir = r'D:\PycharmProjects\pro_dis_2\collected_data\3_combined_images'

# List of specific word image base names to copy (without the extension)
word_images_to_copy = ['cup', 'eat', 'go', 'hot', 'sit']

# Loop through each set folder in the source directory
for set_folder in os.listdir(source_dir):
    set_folder_path = os.path.join(source_dir, set_folder)

    # Only proceed if it's a directory
    if os.path.isdir(set_folder_path):
        # Extract the prefix (e.g., '01', '02', etc.) from the set folder name
        prefix = set_folder.split('_')[1]  # Assumes set folder is named 'set_xx'

        # Create the corresponding set folder in the destination
        destination_set_path = os.path.join(destination_dir, set_folder)
        os.makedirs(destination_set_path, exist_ok=True)
        print(f"\nCreated/checked set folder: {destination_set_path}")

        # Loop through each specified word image and copy it
        for word_image in word_images_to_copy:
            prefixed_word_image = f"{prefix}_{word_image}.png"  # Add the prefix and extension to match the source file name
            image_path = os.path.join(set_folder_path, prefixed_word_image)
            if os.path.isfile(image_path):
                destination_image_path = os.path.join(destination_set_path, prefixed_word_image)
                shutil.copy(image_path, destination_image_path)
                print(f"Copied {prefixed_word_image} from {set_folder_path} to {destination_image_path}")
            else:
                print(f"{prefixed_word_image} not found in {set_folder_path}")
```

**for renaming folders from set_061 to set_460 (augmented datasets)**

```python
import os
import shutil

# Define the source and target directories
source_dir = r"D:\PycharmProjects\pro_dis_2\collected_data\3_augmented_cropped"
target_dir = r"D:\PycharmProjects\pro_dis_2\collected_data\3_augmented_cropped_renamed"

# Create the target directory if it doesn't exist
os.makedirs(target_dir, exist_ok=True)

# Get a sorted list of all folders in the source directory
folders = [f for f in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, f))]
folders.sort()

# Start renaming from `set_061`
start_index = 61

for folder in folders:
    old_path = os.path.join(source_dir, folder)
    new_folder_name = f"set_{start_index:03d}"  # Format the new folder name
    new_path = os.path.join(target_dir, new_folder_name)

    # Copy the folder and its contents to the new directory with the new name
    shutil.copytree(old_path, new_path)
    print(f"Copied and renamed {folder} to {new_folder_name}")

    start_index += 1  # Increment the index

print("\nFolder copying and renaming completed.")
```

**copying augment data specifically to the original Train dataset**

```python
import os
import shutil

# Define the source (augment) and target (Train) directories
augment_dir = r"E:\!special\6_finalimg_augmented_100\Validation"
train_dir = r"E:\!special\7_finalimg_combined_160\Validation"

# Get the list of word folders in the augment directory
word_folders = [f for f in os.listdir(augment_dir) if os.path.isdir(os.path.join(augment_dir, f))]

for word_folder in word_folders:
    augment_word_path = os.path.join(augment_dir, word_folder)
    train_word_path = os.path.join(train_dir, word_folder)

    # Ensure the corresponding folder exists in the Train directory
    if os.path.exists(train_word_path) and os.path.isdir(train_word_path):
        # Copy all files from the augment folder to the Train folder
        for file in os.listdir(augment_word_path):
            source_file = os.path.join(augment_word_path, file)
            destination_file = os.path.join(train_word_path, file)

            # Copy the file
            shutil.copy(source_file, destination_file)
            print(f"Copied {file} from {word_folder} to {train_word_path}")
    else:
        print(f"Warning: Corresponding folder for {word_folder} not found in Train. Skipping.")

print("\nAll images have been copied.")
```

```python
def convert_spaces_to_underscores(sentence):
    return sentence.replace("_", " ")

def capitalize_words(sentence):
    words = sentence.split("_")  # Split on underscores
    capitalized_words = [word.capitalize() for word in words]
    return "_".join(capitalized_words)

def process_sentences(sentences):
    processed_sentences = []
    for sentence in sentences:
        converted = convert_spaces_to_underscores(sentence.strip())  # Clean extra spaces
        # capitalized = capitalize_words(converted)
        processed_sentences.append(converted)
    return processed_sentences

# Example usage with sentences without quotations (entered raw)
input_data = """
Lipreading_using_Convolutional_Neural_Network
Lip-reading_via_Deep_Neural_Networks_using_Hybrid_Visual_Features
Speech_Recognition_Models_are_Strong_Lip-readers
Sub-word_Level_Lip_Reading_with_Visual_Attention
The_Challenge_of_Multispeaker_Lip-reading
Twostage_Visual_Speech_Recognition_for_Intensive_Care_Patients
Visual_Speech_Recognition_Automatic_System_for_Lip_Reading_of_Dutch
"""

# Parse input into a list of sentences
input_sentences = [line.strip() for line in input_data.split("\n") if line.strip()]

# Process sentences
processed_sentences = process_sentences(input_sentences)
for idx, sentence in enumerate(processed_sentences, start=1):
    print(f"{sentence}")
```
